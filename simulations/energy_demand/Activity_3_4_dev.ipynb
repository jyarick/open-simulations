## <font color="#3A78A6"> Section 3A — Linear Calibration (Scale & Shift) </font>

<hr style="
  border: none;
  height: 2px;
  background: linear-gradient(to right, #222, #6FAF7A, #222);
  margin: 10px 0;
">

<p style="margin-top: 0px;">
This tool computes <b>scale</b> and <b>shift</b> values that map one daily energy profile
onto another using a <b>linear calibration</b>.
</p>

<hr style="border: none; height: 1px; background-color: #bbb; margin: 16px 0;">

<p><b>Concept</b></p>

<p>
You start with an existing energy profile <b>$E_0$</b> with full time-series data.
</p>

<p>
You want to create a new profile, <b>$E_1$</b> with a known daily energy range but no full time-series data.
</p>

<p>
The assumption is that <b>$E_1$ has the same shape as $E_0$</b>, but spans a different
minimum and maximum daily energy.
</p>

<hr style="border: none; height: 1px; background-color: #bbb; margin: 16px 0;">

<p><b>Model</b></p>

<p>
$$
E_1(t) = a\,E_0(t) + b
$$
</p>

<ul>
  <li>$a$ — scale factor</li>
  <li>$b$ — vertical shift</li>
</ul>

<hr style="border: none; height: 1px; background-color: #bbb; margin: 16px 0;">

<p><b>What to enter</b></p>

<ul>
  <li>$E_{0,\min}, E_{0,\max}$ — minimum and maximum of the original profile</li>
  <li>$E_{1,\min}, E_{1,\max}$ — estimated minimum and maximum for the new building type</li>
</ul>

<p>
Click <b>Compute Scale + Shift</b> to calculate the values of $a$ and $b$.
</p>

<hr style="border: none; height: 1px; background-color: #bbb; margin: 16px 0;">

<p>
<font color="red"><b>USAGE NOTE:</b></font><br>
This step <b>does not modify any data</b>.  
Parameters are applied in <b>Section&nbsp;3B</b>.
</p>

## <font color="#C96A4A"> Appendix A — Energy vs Power (15-Minute Data) </font>

<hr style="
  border: none;
  height: 2px;
  background: linear-gradient(to right, #222, #C96A4A, #222);
  margin: 10px 0;
">

<p>
The dataset used in this notebook reports <code>*_kWh</code> values as
<b>energy consumed over a fixed time interval</b>, not instantaneous power.
Each entry corresponds to one 15-minute interval.
</p>

<p>
This distinction matters because power is a <i>rate</i>, while energy is an
<i>accumulated quantity</i>.
To compare demand across users, apply transformations, and aggregate loads
on a grid, the notebook works with
<b>average power over each interval</b>.
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Average power over a time interval</b></p>

<p>Average power over a time interval is defined as:</p>

<p>
$$
\bar{P} = \frac{E}{\Delta t}
$$
</p>

<p>
In this notebook, the timestep is fixed:
</p>

<p>
$$
\Delta t = 15\ \text{minutes} = 0.25\ \text{hours}
$$
</p>

<p>
Substituting gives:
</p>

<p>
$$
\bar{P}(\text{kW}) = \frac{E(\text{kWh})}{0.25} = 4E
$$
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Why this conversion is safe here</b></p>

<p>
Because the timestep is constant across the dataset, converting between
interval energy (kWh) and average power (kW) is a simple rescaling and
<b>does not discard information</b>.
This allows the notebook to store data in <b>energy form</b>
while performing analysis in <b>power form</b>,
without changing the underlying time resolution.
</p>


<hr style="
  border: none;
  height: 2px;
  background: linear-gradient(to right, #222, #C96A4A, #222);
  margin: 10px 0;
">

## <font color="#C96A4A"> Appendix B — Linear Scale & Shift Mapping </font>

<hr style="
  border: none;
  height: 2px;
  background: linear-gradient(to right, #222, #C96A4A, #222);
  margin: 10px 0;
">

<p>
Linear calibration is used when you want to create a new energy profile
<b>$E_1(t)$</b> from an existing profile <b>$E_0(t)$</b>,
but with a different minimum and maximum daily energy.
</p>

<p>
This situation arises when you:
</p>

<ul>
  <li>have <b>full time-series data</b> for one building type, and</li>
  <li>only know the <b>expected energy range</b> for another building type.</li>
</ul>

<p>
The assumption is that the two profiles share the same <b>shape</b>,
but differ in <b>scale</b> and <b>baseline</b>.
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Linear model</b></p>

<p>
The mapping is assumed to be linear:
</p>

<p>
$$
E_1(t) = a\,E_0(t) + b
$$
</p>

<p>
where $a$ controls overall magnitude and $b$ shifts the profile vertically.
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Solving for $a$ and $b$</b></p>

<p>
Let:
</p>

<ul>
  <li>$E_{0,\min}, E_{0,\max}$ be the minimum and maximum of the original profile</li>
  <li>$E_{1,\min}, E_{1,\max}$ be the desired minimum and maximum of the new profile</li>
</ul>

<p>
We require the mapping to satisfy:
</p>

<p>
$$
\begin{aligned}
a\,E_{0,\min} + b &= E_{1,\min} \\
a\,E_{0,\max} + b &= E_{1,\max}
\end{aligned}
$$
</p>

<p>
Subtracting the first equation from the second gives:
</p>

<p>
$$
a = \frac{E_{1,\max} - E_{1,\min}}{E_{0,\max} - E_{0,\min}}
$$
</p>

<p>
Substituting back yields:
</p>

<p>
$$
b = E_{1,\min} - a\,E_{0,\min}
$$
</p>

<p>
These two values uniquely define the linear calibration from $E_0$ to $E_1$.
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>What this does — and does not — change</b></p>

<ul>
  <li>This calibration preserves the <b>relative shape</b> of the profile.</li>
  <li>Only the <b>range</b> (minimum and maximum) is modified.</li>
  <li>No smoothing, shifting, or randomness is applied at this stage.</li>
</ul>

<p>
In the notebook workflow, this step computes <b>parameters only</b>.
The actual profile modification occurs later in the
<b>Data Transformation Editor</b>.
</p>

  border: none;
  height: 2px;
  background: linear-gradient(to right, #222, #C96A4A, #222);
  margin: 10px 0;
">

## <font color="#C96A4A"> Appendix C — Data Transformation Editor (What the Code Applies) </font>

<hr style="
  border: none;
  height: 2px;
  background: linear-gradient(to right, #222, #C96A4A, #222);
  margin: 10px 0;
">

<p>
The Data Transformation Editor operates on a single user’s
<b>daily energy profile</b> as a <b>sequential transformation pipeline</b>.
Each step consumes the output of the previous step and produces a new
energy array.
</p>

<p>
This design is intentional: once a transformation is applied, the editor
treats the result as the new baseline for that user.
For this reason, the editor separates <b>previewing</b> transformations from
<b>committing</b> them.
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p>
Let the original daily energy profile be:
</p>

<p>
$$
E_0(t)
$$
</p>

<p>
Each subsequent operation produces a new profile
$E_k(t)$ according to the equations below.
The order shown here matches the implementation and must be preserved.
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Sequential transformation pipeline</b></p>

<p>
Each transformation operates on the output of the previous step.
This ordering is intentional: earlier steps set the large-scale structure,
while later steps add realism without altering that structure.
</p>

<hr style="border: none; height: 0.5px; background-color: gray; margin: 2px 0;">
<ol>
  <li>
    <b>Scale + Shift</b><br>
    Adjust overall magnitude and baseline:
    <p>
    $$
    E_1(t) = a\,E_0(t) + b
    $$
    </p>
    <p>
    This step rescales the profile while preserving its shape.
    The values of $a$ and $b$ may be computed using linear calibration
    or chosen directly for exploratory modeling.
    </p>
  </li>

  <hr style="border: none; height: 0.5px; background-color: gray; margin: 2px 0;">

  <li>
    <b>Noise / Jitter (optional)</b><br>
    Introduce stochastic variability proportional to the signal:
    <p>
    $$
    E_2(t) = E_1(t) + \varepsilon(t)
    $$
    </p>
    <p>
    where $\varepsilon(t)$ is a random perturbation drawn independently
    at each timestep.
    This step exists to model real-world variability and to prevent
    unrealistically identical load shapes.
    </p>
  </li>

  <hr style="border: none; height: 0.5px; background-color: gray; margin: 2px 0;">

  <li>
    <b>Deviation (contrast control)</b><br>
    Adjust peak–valley contrast about the daily mean:
    <p>
    $$
    E_3(t) = \mu + f\,[E_2(t) - \mu]
    $$
    </p>
    <p>
    where $\mu$ is the mean of $E_2(t)$.
    This rescales both structure and variability relative to the mean.
    </p>
  </li>

  <hr style="border: none; height: 0.5px; background-color: gray; margin: 2px 0;">

  <li>
    <b>Hourly time weights</b><br>
    Apply hour-specific multipliers:
    <p>
    $$
    E_4(t) = w_h\,E_3(t)
    $$
    </p>
    <p>
    where $w_h$ is the weight for the hour containing timestep $t$
    (each hour spans four 15-minute bins).
    </p>
  </li>

  <hr style="border: none; height: 0.5px; background-color: gray; margin: 2px 0;">

  <li>
    <b>Smoothing (optional)</b><br>
    Reduce high-frequency variation via Gaussian smoothing:
    <p>
    $$
    E_5(t) = \mathcal{G}\!\left[E_4(t)\right]
    $$
    </p>
    <p>
    This step represents physical inertia or aggregated behavior over time
    and is applied only if SciPy is available.
    </p>
  </li>

  <hr style="border: none; height: 0.5px; background-color: gray; margin: 2px 0;">

  <li>
    <b>Time shift</b><br>
    Circularly shift the profile by an integer number of 15-minute intervals:
    <p>
    $$
    E_6(t) = E_5(t - \Delta t_{\text{shift}})
    $$
    </p>
  </li>

  <hr style="border: none; height: 0.5px; background-color: gray; margin: 2px 0;">

  <li>
    <b>Final clipping (physical validity)</b><br>
    Enforce non-negative energy values after all transformations:
    <p>
    $$
    E_7(t) = \max\!\left(E_6(t),\,0\right)
    $$
    </p>
    <p>
    This step exists solely to ensure physical validity.
    It does not change the intended profile shape.
    </p>
  </li>
</ol>

<hr style="border: none; height: 1px; background-color: gray; margin: 10px 0;">

<p>
The final transformed profile is $E_7(t)$.
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Noise, seeding, and reproducibility</b></p>

<p>
Noise is the only <b>stochastic</b> step in the pipeline.
Because it changes numerical outcomes, the editor explicitly separates
<b>variability</b> from <b>reproducibility</b>.
</p>

<ul>
  <li>
    When <b>Lock Noise (Reproducible)</b> is enabled, a fixed random seed is used.
    Applying the same parameters produces the same noise realization
    and therefore the same downstream profiles.
  </li>

  <li>
    When noise is <b>not locked</b>, a new random realization is generated
    each time Apply is clicked.
    In this mode, the seed value is intentionally ignored.
  </li>
</ul>

<p>
Randomness should only be “frozen” <b>after</b> a student decides
a specific scenario should be preserved.
Before that point, changing noise realizations is expected behavior.
</p>

<p>
Locking noise supports:
</p>

<ul>
  <li><b>Controlled experiments</b> and fair grading</li>
  <li><b>Reproducibility</b> across exports and saved results</li>
</ul>

<p>
Unlocked noise supports:
</p>

<ul>
  <li><b>Exploration</b> of stochastic behavior</li>
  <li><b>Realistic diversity</b> between otherwise similar users</li>
</ul>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Preview vs Apply</b></p>

<p>
<b>Preview</b> computes the full transformation pipeline
$E_0 \rightarrow E_7$ without modifying stored state.
<b>Apply</b> commits the final profile and replaces the user’s active dataset.
</p>

<p>
Because the pipeline is sequential and may include randomness,
this separation prevents accidental drift and ensures that
committed results are intentional.
</p>

## <font color="#C96A4A"> Appendix D — Power Computation & Aggregation </font>

<hr style="
  border: none;
  height: 2px;
  background: linear-gradient(to right, #222, #C96A4A, #222);
  margin: 10px 0;
">

<p>
Power is computed <b>only after all energy transformations are finalized</b>,
using the final per-building energy profiles and the specified number of buildings
for each user type.
</p>

<p>
This separation is intentional: transformations modify the
<b>shape</b> of a single building’s daily energy profile,
while aggregation controls overall <b>scale</b> through building counts.
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Per-user power computation</b></p>

<p>
For a single user type, average power at each timestep is computed as:
</p>

<p>
$$
\bar{P}_{\text{user}}(t)
= 4\,E_{\text{user}}(t)\times N_{\text{buildings}}
$$
</p>

<p>
where:
</p>

<ul>
  <li>$E_{\text{user}}(t)$ is the final transformed <b>per-building</b> energy profile,</li>
  <li>$N_{\text{buildings}}$ is the confirmed building count, and</li>
  <li>the factor of 4 converts 15-minute interval energy (kWh)
      to average power (kW).</li>
</ul>

<p>
This calculation is performed independently at each 15-minute timestep.
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Total Grid aggregation</b></p>

<p>
The <b>Total Grid</b> power profile is computed as the sum across all selected users:
</p>

<p>
$$
\bar{P}_{\text{Total}}(t) = \sum_i \bar{P}_i(t)
$$
</p>

<p>
Aggregation affects only the <b>magnitude</b> of demand.
No reshaping, smoothing, or timing adjustments occur at this stage;
all temporal structure is inherited directly from the transformed
energy profiles.
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Peak percentage interpretation</b></p>

<p>
The percentage column answers the question:
</p>

<p>
<i>
How does this user's peak demand compare to the grid's peak demand?
</i>
</p>

<p>
Formally:
</p>

<p>
$$
\text{Percent of Total Grid Peak}
=
100 \times
\frac{P_{\max,\ \text{user}}}
     {P_{\max,\ \text{Total Grid}}}
$$
</p>

<p>
where $P_{\max,\ \text{Total Grid}}$ is the grid’s peak demand
at the single timestep when the Total Grid reaches its maximum.
</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Conceptual understanding</b></p>

<p>
These percentages generally <b>do not sum to 100%</b>.
This is expected behavior.
</p>

<p>
Individual users typically reach their peak demand at
<b>different times</b> than the grid’s peak.
The percentages shown compare each user’s
<b>own maximum demand</b> to the
<b>maximum demand of the total grid</b>,
even though those peaks may occur at different times.
</p>

<p>
In the special case where all users peak at the same time,
the percentages would sum to 100%.
</p>

<p>
In practice, the sum of percentages can approach 100% when one or more
<b>dominant user types</b> are present.
A user may be dominant because:
</p>

<ul>
  <li>it has a very large per-building demand, or</li>
  <li>it exists in large numbers, causing moderate individual demand
      to aggregate into a large grid contribution.</li>
</ul>

<p>
Importantly, percentages are <b>not</b> normalized against the sum of all
users’ individual peak values.
They are normalized against the
<b>maximum demand ever reached by the Total Grid</b>,
even though individual users may peak at different times.
</p>

## <font color="#C96A4A"> Appendix F — Time Indexing, Windows, & Circular Shifts </font>

<hr style="
  border: none;
  height: 2px;
  background: linear-gradient(to right, #222, #C96A4A, #222);
  margin: 10px 0;
">

<p>
This appendix explains how <b>time indexing</b>, <b>time windows</b>,
and <b>circular shifts</b> work in this notebook.
</p>

<p>
These details are not required to complete the activity,
but they are useful if you want to understand
<b>why profiles move the way they do</b>
and how timing affects aggregation.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Fixed timestep</b></p>

<p>
All data in this notebook uses a <b>fixed 15-minute timestep</b>.
Each day is divided into:
</p>

<p>
$$
96 \text{ timesteps per day}
$$
</p>

<p>
Time is therefore indexed as:
</p>

<p>
$$
t = 0, 1, 2, \dots, 95
$$
</p>

<p>
Each index corresponds to a specific 15-minute interval,
not a continuously varying clock time.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Time windows</b></p>

<p>
The notebook allows analysis over a selected <b>time window</b>,
which may represent:
</p>

<ul>
  <li>a single day,</li>
  <li>a partial day, or</li>
  <li>a longer continuous slice of data.</li>
</ul>

<p>
Changing the time window affects:
</p>

<ul>
  <li>which timesteps are plotted,</li>
  <li>which timesteps are exported, and</li>
  <li>where peaks may appear.</li>
</ul>

<p>
It does <b>not</b> change the meaning of individual timesteps,
the energy-to-power conversion,
or the transformation math.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Circular time shifts</b></p>

<p>
When a time shift is applied in the Data Transformation Editor,
the profile is shifted <b>circularly</b>.
</p>

<p>
This means values shifted past the end of the day
wrap around to the beginning.
</p>

<p>
Mathematically:
</p>

<p>
$$
E(t) \;\rightarrow\; E(t - \Delta t_{\text{shift}})
$$
</p>

<p>
with indexing performed modulo 96.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Physical interpretation</b></p>

<p>
A circular shift represents a <b>change in timing</b>,
not a change in total energy.
</p>

<ul>
  <li>Total daily energy is preserved.</li>
  <li>Peak timing may move earlier or later.</li>
  <li>Relative alignment between users may change.</li>
</ul>

<p>
This is intentional: shifting allows students to explore
how <b>coincidence of demand</b> affects grid-level peaks
without altering per-building usage.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Shifts and aggregation</b></p>

<p>
When multiple users are aggregated:
</p>

<ul>
  <li>shifting a single user can <b>increase</b> or <b>decrease</b>
      the Total Grid peak,</li>
  <li>even if that user’s total energy remains unchanged.</li>
</ul>

<p>
This occurs because grid peaks depend on
<b>simultaneous demand</b>, not individual maxima.
</p>

<p>
Time shifts therefore affect <b>when</b> demand occurs,
not <b>how much</b> energy is used.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Why circular shifts are used</b></p>

<p>
Circular shifting avoids edge effects that would otherwise:
</p>

<ul>
  <li>introduce artificial zeros,</li>
  <li>truncate demand near the boundaries, or</li>
  <li>change total energy.</li>
</ul>

<p>
This makes timing experiments cleaner and easier to interpret,
especially when comparing multiple users.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Summary</b></p>

<ul>
  <li>Time is discretized into fixed 15-minute intervals.</li>
  <li>Time windows select which intervals are analyzed.</li>
  <li>Time shifts move demand <b>without</b> changing total energy.</li>
  <li>Circular wrap-around preserves physical consistency.</li>
</ul>

<p>
These design choices allow timing effects to be explored
independently from energy magnitude.
</p>

## <font color="#C96A4A"> Appendix G — Model Assumptions & Limits </font>

<hr style="
  border: none;
  height: 2px;
  background: linear-gradient(to right, #222, #C96A4A, #222);
  margin: 15px 0;
">

<p>
This appendix summarizes the <b>assumptions</b> built into the model
and the <b>limits</b> of the conclusions that can be drawn from it.
</p>

<p>
The goal is not to list flaws, but to clarify
<b>what the model is designed to represent</b>
and what is intentionally left out.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Core assumptions</b></p>

<ul>
  <li>
    Each building type is represented by a
    <b>typical daily energy profile</b>.
  </li>

  <li>
    That profile is assumed to repeat for the selected time window
    unless explicitly modified.
  </li>

  <li>
    All buildings of the same user type share the same
    underlying profile shape.
  </li>

  <li>
    Energy use within each 15-minute interval is treated
    as uniformly distributed over that interval.
  </li>
</ul>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Aggregation assumptions</b></p>

<ul>
  <li>
    Total Grid demand is computed as a
    <b>simple sum</b> of individual user demands.
  </li>

  <li>
    There are no feedback effects between users
    (one user’s behavior does not alter another’s).
  </li>

  <li>
    Building counts scale demand linearly.
  </li>
</ul>

<p>
These assumptions isolate the effects of
<b>profile shape</b>, <b>timing</b>, and <b>count</b>
without introducing additional system complexity.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>What this model does not include</b></p>

<ul>
  <li>Weather variation or seasonal effects</li>
  <li>Day-to-day behavioral changes</li>
  <li>Demand response or load shedding</li>
  <li>Physical grid constraints (lines, transformers, losses)</li>
  <li>Electricity pricing or operational cost</li>
</ul>

<p>
As a result, the model should be interpreted as a
<b>demand-shape exploration tool</b>,
not a full power system simulator.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>How to interpret results</b></p>

<ul>
  <li>
    Changes in <b>peak power</b> reflect timing and coincidence effects,
    not infrastructure limits.
  </li>

  <li>
    Changes in <b>total energy</b> reflect scaling and profile magnitude,
    not efficiency improvements.
  </li>

  <li>
    Two scenarios with the same total energy
    can produce very different peak demands.
  </li>
</ul>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Appropriate conclusions</b></p>

<ul>
  <li>
    Which building types tend to drive grid peaks
  </li>

  <li>
    How timing alignment affects Total Grid demand
  </li>

  <li>
    How scaling certain users changes peak stress
  </li>
</ul>

<p>
These insights are sufficient for comparing scenarios
and understanding tradeoffs in community-level demand.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Summary</b></p>

<ul>
  <li>This notebook models <b>demand behavior</b>, not grid physics.</li>
  <li>Assumptions are intentional and simplify interpretation.</li>
  <li>Results should be used for <b>comparison</b>, not prediction.</li>
</ul>

<p>
Understanding these limits helps ensure that conclusions
drawn from the model are accurate and appropriate.
</p>

## <font color="#C96A4A"> Appendix H — Extending This Notebook Toward Machine Learning </font>

<hr style="
  border: none;
  height: 2px;
  background: linear-gradient(to right, #222, #C96A4A, #222);
  margin: 10px 0;
">

<p>
This appendix outlines how the ideas in this notebook
could be extended toward <b>machine learning (ML)</b>.
</p>

<p>
ML is <b>not required</b> for this activity.
This section is included for students who are curious about
how modeling, simulation, and data-driven methods connect.
</p>

<p>⚠️ Machine learning is introduced here not to replace physical calculations, but to approximate and invert them when the design space becomes too large to explore directly.</p>

<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>What this notebook already provides</b></p>

<p>
By the time you reach the end of this activity, you have already built:
</p>

<ul>
  <li>clean, aligned <b>time-series data</b>,</li>
  <li>controlled <b>scenario variations</b>, and</li>
  <li>clear <b>numerical outcomes</b> (peaks, timing, contributions).</li>
</ul>

<p>
This is exactly the kind of structured, labeled data
that machine learning models require.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Compare vs predict</b></p>

<p>
The goal of extending this notebook toward ML would be to
<b>compare scenarios</b>, not to predict the real future.
</p>

<p>
For example:
</p>

<ul>
  <li>Which scenario produces a higher grid peak?</li>
  <li>Which user types most strongly influence peak demand?</li>
  <li>How does timing matter relative to magnitude?</li>
</ul>

<p>
This differs from prediction, which would attempt to forecast
actual future demand using real-world data.
</p>

<p>
The notebook is intentionally designed for comparison and exploration,
not for real-time forecasting.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Learning from scenarios (supervised learning)</b></p>

<p>
A natural first ML step would be to generate many valid scenarios
using this notebook and train a model to learn relationships between:
</p>

<ul>
  <li><b>Inputs</b>: user counts, transformation parameters, time shifts</li>
  <li><b>Outputs</b>: grid peak power, peak timing, user contributions</li>
</ul>

<p>
The model would not predict the future.
It would learn how different design choices
<b>compare</b> to one another.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>What the model would learn</b></p>

<ul>
  <li>Which parameters most strongly drive peak demand</li>
  <li>Nonlinear interactions between users</li>
  <li>How synchronization affects grid stress</li>
</ul>

<p>
This turns the notebook into a <b>data generator</b>
and the ML model into a fast <b>surrogate</b>
for exploring many scenarios.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Learning demand shapes (advanced)</b></p>

<p>
Instead of predicting a single value like peak power,
an ML model could be trained to predict the
<b>entire grid power curve</b>.
</p>

<p>
In this case:
</p>

<ul>
  <li>Inputs describe the scenario setup</li>
  <li>Outputs are the full time-series demand profile</li>
</ul>

<p>
This approach focuses on timing, coincidence, and shape,
not just magnitude.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Inverse design</b></p>

<p>
An advanced extension is to reverse the problem:
</p>

<p>
<i>
“What changes would reduce the grid peak by a desired amount?”
</i>
</p>

<p>
This involves:
</p>

<ul>
  <li>a trained model that estimates grid response, and</li>
  <li>an optimization step that searches for better scenarios.</li>
</ul>

<p>
This mirrors how ML is used in real engineering design,
not as a prediction tool, but as a guide for decision-making.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Toward real-world forecasting (optional)</b></p>

<p>
The machine learning approaches described above
learn relationships within this notebook’s model.
They do <b>not</b> predict real future demand.
</p>

<p><b><font color="purple">Question:</font> What data would be needed for real forecasting?</b></p>

<p>
To predict future demand (instead of comparing scenarios),
an ML model needs <b>historical ground-truth data</b>
and variables that drive demand.
</p>

<ul>
  <li><b>Historical demand</b>: measured power (kW) or energy (kWh) over many days</li>
  <li><b>Weather</b>: temperature (most important), humidity, extreme events</li>
  <li><b>Calendar</b>: weekday/weekend, holidays, school schedules</li>
  <li><b>Community context</b>: population changes, special events, outages (if available)</li>
</ul>

<p>
With those inputs, forecasting becomes a problem of learning
patterns in real data — not just learning behavior inside a simulator.
</p>

<p>
This distinction is intentional:
learning how design choices affect demand
is a necessary step <b>before</b> attempting prediction.
</p>

<p><b><font color="purple">Question:</font> How would you judge a forecast?</b></p>

<p>
A forecast is only useful if it performs well on data it has never seen.
This is tested by training on earlier days and evaluating on later days.
</p>

<ul>
  <li><b>MAE</b>: average absolute error (typical “how far off”)</li>
  <li><b>Peak error</b>: how accurate the model is on the highest-demand periods</li>
  <li><b>Timing error</b>: how close the predicted peak time is to the real peak time</li>
</ul>

<p>
In power systems, getting peak magnitude and peak timing right
often matters more than being perfect everywhere.
</p>


<hr style="border: none; height: 1px; background-color: gray; margin: 5px 0;">

<p><b>Summary</b></p>

<ul>
  <li>This notebook already generates <b>ML-ready, structured data</b>.</li>
  <li>Machine learning can be used to <b>compare scenarios</b>
      and understand how design choices affect demand.</li>
  <li>Predicting real future demand would require
      <b>additional real-world data</b>
      and is intentionally out of scope here.</li>
</ul>

<p>
Seen this way, the notebook is both a
<b>learning tool</b> and a <b>foundation</b>
for more advanced modeling.
</p>
